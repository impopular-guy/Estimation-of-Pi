# Monte Carlo Estimation of Pi

$\pi$ approximation using Monte Carlo method is a probabilistic approach where the value of $\pi$ is obtained by calculating the number of random points that fall on the specified region.

Let's inscribe a circle of radius R with centre as Origin in a square of side length 2R. For simplicity we will consider only the first quadrant. If we randomly pick a point inside the square the probability that the point lies inside the circle will be the ratio of area of circle to the area of square.
$$P = {{area\space of {1\over{4}}th\space circle}\over{area\space of {1\over{4}}th}\space square}$$
this implies,$$P = {\pi R^2\over (2R)^2} = {\pi\over 4}$$
Let us pick $N$ random points out of which $M$ lies inside the circle. Hence,$$P = {M\over N}$$
this implies,$$ \pi = {4M\over N} $$

A point $(x, y)$ will inside the circle if $x^2 + y^2 < R^2$

If $R = 1$, them we can say $x^2 + y^2 <1$

Let's not waste anymore time and jump straight to the code.

```python
#import the libraries
from random import random as rn
import math as m
```

```python
N = 10  #total no. of points

print('{:9} {:9}'.format('N','Pi'))
for a in range(2,9):
    
    inside = 0 #no. of points inside circle
    for i in range(N):
        if ( rn()**2 + rn()**2 ) < 1:
            inside = inside+1
    
    pi = 4*inside/N
    print('%-8d %-8f'%(N, pi))
    
    N=10**a #updating no.of points
```

**Graphical Representation**

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import numpy as np
xy = np.array(np.random.random((1000,2)))
xy_in = np.array([p for p in xy if p[0]**2 + p[1]**2 < 1 ])
xy_out = np.array([p for p in xy if p[0]**2 + p[1]**2 > 1 ])
plt.gca().add_patch(Arc( (0,0), 2,2, theta1 = 0.0, theta2 = 90.0) )
plt.scatter(xy_in[:,0], xy_in[:,1], color = 'red',s=5)
plt.scatter(xy_out[:,0], xy_out[:,1], color = 'blue', s=5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axes().set_aspect(1)
plt.xlim(0,1)
plt.ylim(0,1)
plt.title('Plotting 1000 random points')
plt.show()
```

**Coclusion**

We achieved correct reading only upto two decimal places. This method doesn't have a fast convergence. Monte Carlo approximation is a slow method for approximation of $\pi$.
